```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
```

```python
df=pd.read_csv('pima_indian.csv')
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
features_col=['num_preg','glucose_conc','diastolic_bp','thickness','insulin','bmi','diab_pred','age']
class_name=['diabetes']
```

```python
X = df[features_col].values
X
```

```python
y=df[class_name].values
y
```

```python
#Split data into training and testing data
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
ngb=GaussianNB().fit(X_train,y_train.ravel())
```

```python
predicted=ngb.predict(X_test)
predicted
```

```python
newData=ngb.predict([[6,148,72,35,0,33.6,0.627,50]])
newData
```

```python
print("Confusion Matrix:")
print(metrics.confusion_matrix(y_test,predicted))
```

```python
print("Accuracy:",metrics.accuracy_score(y_test,predicted))
```

```python
print("Precision:",metrics.precision_score(y_test,predicted))
print("Recall:",metrics.recall_score(y_test,predicted))
```
